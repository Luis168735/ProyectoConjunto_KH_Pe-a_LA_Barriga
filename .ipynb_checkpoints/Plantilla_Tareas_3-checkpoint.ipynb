### Código de solución estudiante 2

```python
### 1 Resolver la siguiente ecuacion recursiva usando funciones 
## D1=0 y D2=1
def D(n):
    if n == 1:
        return 0
    elif n == 2:
        return 1
    else:
        return (n-1)*D(n-1)+(n-1)*D(n-2)
```

```python
# D1 y D2
D(1),D(2)
```

```python
# n mayor o igual a 3
print([D(n) for n in range(3,10)])
```

```python
### 4 Find all of the words in a string that are less than 4 letters 
def find_short_words(x):
    words = x.split()
    short_words = []
    for word in words:
        if len(word) < 4:
            short_words.append(word)
    return short_words
```

7. Como ejemplo simple de una simulación de Monte Carlo, considere calcular la probabilidad de una suma particular del lanzamiento de tres dados (cada dado tiene valores del uno al seis). Además cada dado tiene las siguientes carácterísticas: el primer dado no está cargado (distribución uniforme todos son equiprobables); el segundo y tercer dado están cargados basados en una distribución binomial con parámetros (`n=5, p=0.5` y `n=5, p=0.2`). Calcule la probabilidad de que la suma resultante sea 7, 14 o 18. 

```python
binom.pmf?
```

```python
import numpy as np
from scipy.stats import binom

def sum_probability(sum_value):
    p_uniform = 1/6  # probability of getting a specific value on the uniform dice
    p_binom_1 = binom.pmf(sum_value, 5, 0.5)  # probability of getting the sum value on the binomial dice
    p_binom_2 = binom.pmf(sum_value, 5, 0.2)
    
    # calculate the probability of getting the specified sum value by summing over all possible outcomes
    prob = 0
    for i in range(1, 7):
        for j in range(1, 7):
            for k in range(1, 7):
                if i+j+k == sum_value:
                    prob += p_uniform + p_binom_1 + p_binom_2

    return prob
```

```python
print([sum_probability(sum_value) for sum_value in range(6,19)])
```
